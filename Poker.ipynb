```python
class Card:

    def __init__(self, rank=0, suit=0):
        ranks = [2,3,4,5,6,7,8,9,10,"Jack","Queen","King","Ace"]
        suits = ["Clubs","Diamonds","Hearts","Spades"]
        self.rank = ranks[rank]
        self.suit = suits[suit]

    def __str__ (self):
        return "{0} of {1}".format (self.rank, self.suit)
```

```python
class Hand:   

    def __init__ (self, name= " ", card1=Card(0,0), card2=Card(0,0)):
        self.name = name
        self.card1 = card1
        self.card2 = card2

    def __str__ (self):
        return "{0} is having {1}, {2}".format (self.name, self.card1, self.card2)
```

```python
class Stacks:
    
        def __init__ (self, stacks={}):
            self.stacks = stacks

        def __str__ (self):
            for stack in self.stacks:
                print (stack.name+"'s stack"+":", self.stacks[stack],"bb")
            return "___________________\n"
        
        def delete_stack (self, name):
            del self.stacks[name]
```

```python
class Deck (Stacks):

    def __init__ (self):
        import random
        rng = random.Random()
        ranks = [2,3,4,5,6,7,8,9,10,"Jack","Queen","King","Ace"]
        suits = ["Clubs","Diamonds","Hearts","Spades"]
        self.deck = []
        for suit in range (len (suits)):
            for rank in range (len (ranks)):
                card = Card (rank, suit)
                self.deck.append (card)
        rng.shuffle (self.deck)

    def __str__ (self):
        print ("The deck is:\n")
        for card in self.deck:
            print (card)
        return "___________________\n"

    def pop_card (self):
        return self.deck.pop()
```
